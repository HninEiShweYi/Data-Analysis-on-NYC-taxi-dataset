```python
import matplotlib.pyplot as plt
import numpy as np

#Variation in Trip Volume
trip_counts = [len(df_january), len(df_march), len(df_june)]
months = ["January", "March", "June"]

print("1. Variation in Trip Volume:")
for i, month in enumerate(months):
    print(f"{month}: {trip_counts[i]} trips")


plt.figure(figsize=(8, 6))
plt.bar(months, trip_counts, alpha=0.7, color='b')
plt.title("Variation in Trip Volume")
plt.xlabel("Month")
plt.ylabel("Number of Trips")
plt.show()
```

```python
# Calculate the average fare amount for each month
# Calculate the average fare amount for each month
average_fare_january = df_january['fare_amount'].mean()
average_fare_march = df_march['fare_amount'].mean()
average_fare_june = df_june['fare_amount'].mean()

# Create a bar plot to visualize the trend of average fare amount
months = ["January", "March", "June"]
average_fares = [average_fare_january, average_fare_march, average_fare_june]

plt.figure(figsize=(8, 6))
bars = plt.bar(months, average_fares, alpha=0.7, color='b')
plt.title("Average Fare Amount Comparison")
plt.xlabel("Month")
plt.ylabel("Average Fare Amount")

# Add text labels with the average fare amounts on top of the bars
for bar, avg_fare in zip(bars, average_fares):
    plt.text(bar.get_x() + bar.get_width()/2 - 0.15, bar.get_height() + 0.5, f"{avg_fare:.2f}", ha='center', color='black', fontsize=12)

plt.show()
```

```python
# Calculate the total tip amount for each month
total_tip_january = df_january['tip_amount'].sum()
total_tip_march = df_march['tip_amount'].sum()
total_tip_june = df_june['tip_amount'].sum()

# Create a bar plot to visualize the comparison of total tip amounts
months = ["January", "March", "June"]
total_tips = [total_tip_january, total_tip_march, total_tip_june]

plt.figure(figsize=(8, 6))
plt.bar(months, total_tips, alpha=0.7, color='g')
plt.title("Comparison of Total Tip Amounts")
plt.xlabel("Month")
plt.ylabel("Total Tip Amount")

# Add text labels with the total tip amounts on top of the bars
for i, tip_amount in enumerate(total_tips):
    plt.text(i, tip_amount, f"${tip_amount:.2f}", ha='center', va='bottom', fontsize=12)

plt.show()
```

```python
# Calculate the total count of each payment type for each month
january_payment_counts = df_january['payment_type'].value_counts().reindex([1, 2, 3, 4, 5, 6], fill_value=0)
march_payment_counts = df_march['payment_type'].value_counts().reindex([1, 2, 3, 4, 5, 6], fill_value=0)
june_payment_counts = df_june['payment_type'].value_counts().reindex([1, 2, 3, 4, 5, 6], fill_value=0)

# Create a bar plot to visualize the comparison of payment types
payment_labels = ["Credit card", "Cash", "No charge", "Dispute", "Unknown", "Voided trip"]
x = np.arange(len(payment_labels))

plt.figure(figsize=(10, 6))

plt.bar(x - 0.2, january_payment_counts, alpha=0.7, width=0.2, label="January")
plt.bar(x, march_payment_counts, alpha=0.7, width=0.2, label="March")
plt.bar(x + 0.2, june_payment_counts, alpha=0.7, width=0.2, label="June")

plt.title("Payment Type Comparison")
plt.xlabel("Payment Type")
plt.ylabel("Count")
plt.legend()
plt.xticks(x, payment_labels, rotation=45)

# Print the payment type counts for each month
print("Payment Type Comparison:")
print("Payment Type\tJanuary\tMarch\tJune")
for label, count_january, count_march, count_june in zip(payment_labels, january_payment_counts, march_payment_counts, june_payment_counts):
    print(f"{label}\t{count_january}\t{count_march}\t{count_june}")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Assuming you have already loaded your DataFrame, let's call it df_january

# Define the rush hour time periods (you can adjust these as needed)
morning_rush_start = 7  # Morning rush hour starts at 7:00 AM
morning_rush_end = 9    # Morning rush hour ends at 9:00 AM
evening_rush_start = 17  # Evening rush hour starts at 5:00 PM
evening_rush_end = 19    # Evening rush hour ends at 7:00 PM

# Create a function to determine if a trip is during rush hour
def is_rush_hour(pickup_datetime):
    pickup_hour = pickup_datetime.hour
    if (morning_rush_start <= pickup_hour < morning_rush_end) or (evening_rush_start <= pickup_hour < evening_rush_end):
        return 1  # Trip is during rush hour
    else:
        return 0  # Trip is not during rush hour

# Apply the function to create the rush-hour flag column
df_january['rush_hour_flag'] = df_january['tpep_pickup_datetime'].apply(is_rush_hour)

# Display the first few rows of the DataFrame with the new rush-hour flag column
print(df_january[['tpep_pickup_datetime', 'rush_hour_flag']].head())




```

```python
import numpy as np

# Calculate the straight-line distance (Euclidean distance) for each trip
df_january['straight_line_distance'] = np.sqrt(
    (df_january['PULocationID'] - df_january['DOLocationID']) ** 2
)

# Calculate the ratio of actual distance to straight-line distance
df_january['distance_complexity'] = df_january['trip_distance'] / df_january['straight_line_distance']

# 'distance_complexity' column will now contain the complexity ratio for each trip


```

```python
# Calculate pickup frequency in each taxi zone
pickup_frequency = df_january['PULocationID'].value_counts().reset_index()
pickup_frequency.columns = ['TaxiZone', 'PickupFrequency']

# Calculate drop-off frequency in each taxi zone
dropoff_frequency = df_january['DOLocationID'].value_counts().reset_index()
dropoff_frequency.columns = ['TaxiZone', 'DropoffFrequency']

# Merge the pickup and drop-off frequencies based on the 'TaxiZone' column
taxi_zone_frequencies = pickup_frequency.merge(dropoff_frequency, on='TaxiZone', how='outer')

# Fill any missing values with 0 (for zones with either pickup or dropoff but not both)
taxi_zone_frequencies.fillna(0, inplace=True)

# Display the resulting DataFrame
print(taxi_zone_frequencies)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have already loaded df_january from your dataset

# Extract the hour from the pickup datetime and create a new column
df_january['pickup_hour'] = df_january['tpep_pickup_datetime'].dt.hour

# Group the data by pickup hour and count the number of trips for each hour
hourly_trip_counts = df_january['pickup_hour'].value_counts().sort_index()

# Plot the hourly trip counts
plt.figure(figsize=(12, 6))
plt.bar(hourly_trip_counts.index, hourly_trip_counts.values, alpha=0.7, color='b')
plt.title("Peak Travel Hours")
plt.xlabel("Hour of the Day")
plt.ylabel("Number of Trips")
plt.xticks(hourly_trip_counts.index)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Find the peak travel hours
peak_hours = hourly_trip_counts[hourly_trip_counts == hourly_trip_counts.max()]
print("Peak Travel Hours:")
print(peak_hours)

```

```python
# Assuming you have already loaded df_january from your dataset

# Extract the day of the week from the pickup datetime (0 = Monday, 6 = Sunday)
df_january['pickup_day_of_week'] = df_january['tpep_pickup_datetime'].dt.dayofweek

# Define a function to categorize days as weekday or weekend
def categorize_day(day):
    if day < 5:  # Monday to Friday are weekdays (0 to 4)
        return 'Weekday'
    else:  # Saturday and Sunday are weekends (5 and 6)
        return 'Weekend'

# Apply the categorization function to create a new column
df_january['day_category'] = df_january['pickup_day_of_week'].apply(categorize_day)

# Group the data by day category and calculate the average trip distance for each category
average_distance_by_day = df_january.groupby('day_category')['trip_distance'].mean()

# Print the average distances for weekdays and weekends
print("Average Distance on Weekdays:", average_distance_by_day['Weekday'])
print("Average Distance on Weekends:", average_distance_by_day['Weekend'])
```

The average distance of taxi trips on weekdays is approximately 3.74 miles, while the average distance on weekends is approximately 4.11 miles. This suggests that taxi trips tend to be slightly longer on weekends compared to weekdays.
